```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a dictionary of the player data for this first viz
data = {'Player':['Duncan', 'Manu', 'Parker', 'Lebrawn Jaims', 'Kobe', 'Yao', 'Jokic', 'Shaq'], 
        'PPG':[19.0,13.3,15.5,27.1,25,19,19.7,23.7], 
        'WS_48':[.209,.19,.14,.23,.17,.20,.239,.208]} 

df = pd.DataFrame(data)
df.set_index('Player', inplace=True)
df
```

```python
fig, ax = plt.subplots()
ax.scatter(df['PPG'], df['WS_48']);
```

```python
# Regular season data

X=df['PPG']
y=df['WS_48']

plt.figure(figsize=(16,10))
plt.scatter(X, y, s=50,cmap='Set1')
plt.suptitle('Win Shares per 48 mins by PPG', fontsize=36)
plt.title('Cool Players Only', fontsize=28);
plt.xlabel('PPG',fontsize=20)
plt.ylabel('Win Shares/48 mins',fontsize=20)
plt.tick_params(axis="x", labelsize=18)
plt.tick_params(axis="y", labelsize=18)
for i in range(df.shape[0]):
    plt.text(x=df.PPG[i]+0.3,y=df.WS_48[i],s=df.index[i], 
          fontdict=dict(color='blue',size=15),rotation=35)
```
